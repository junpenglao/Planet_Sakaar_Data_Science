```python
import pymc3 as pm
import numpy as np
```

```python
with pm.Model() as m:
    JSkill = pm.Normal('JSkill', 120., 40.)
    FSkill = pm.Normal('FSkill', 100., 5.)
    lg_p = pm.distributions.dist_math.normal_lccdf(JSkill - FSkill, pm.math.sqrt(50.), 0.)
    JWin = pm.Bernoulli('JWin', pm.math.exp(lg_p), observed=1.)
    trace = pm.sample()
```

```python
pm.traceplot(trace);
```
